```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np
import csv
%matplotlib inline
```

```python
nba = pd.read_csv('AllStarNBA')
nba_2016=pd.read_csv('NBA_2016')
```

```python
nba.info()
```

```python
nba.team= np.where(nba.team != 'None',1,0)
```

```python
plt.title("Distribution of Honorees, (1 = Made Team)")  
nba.team.value_counts().plot(kind='bar')
```

```python
alpha_level = .35
nba.pts[nba.team ==0].plot(kind = 'hist',alpha=alpha_level)
nba.pts[nba.team ==1].plot(kind = 'hist')
plt.xlabel("Pts")
plt.ylabel("Points Distribution" )
plt.legend(('No team','All NBA Team', ),loc='best')
```

```python
nba.WS[nba.team ==1].plot(kind = 'hist')
nba.WS[nba.team ==0].plot(kind = 'hist', alpha=alpha_level)
plt.legend(('All NBA Team', 'No team'),loc='best')
plt.xlabel("Win Shares")
plt.ylabel("Win Share Distribution" )
```

```python
nba.mp[nba.team ==1].plot(kind = 'hist')
nba.mp[nba.team ==0].plot(kind = 'hist',alpha=alpha_level)
plt.legend(('All NBA Team', 'No team'),loc='best')
plt.xlabel("Minutes Played")
plt.ylabel("Minutes Played Distribution" )
```

```python
nba.PER[nba.team ==1].plot(kind = 'kde')
nba.PER[nba.team ==0].plot(kind = 'kde',alpha=alpha_level)
plt.legend(('All NBA Team', 'No team'),loc='best')
plt.xlabel("Player Efficiency")
plt.ylabel("Player Efficiency Rating Distribution" )
```

```python
from pandas.tools.plotting import parallel_coordinates
features = [["g","gs","mp","fg","fga","fg_","x3p","x3pa","x3p_","x2p","x2pa","x2p_","ft","fta","ft_","orb","drb","trb","ast","stl","blk","tov","pf","pts"]]
#features = [['g','gs','mp','pts','AST%','PER','STL%','USG%','FTr','3PAr','TS%','VORP','BPM','OBPM','DBPM','WS','WS/48','DWS','OWS',]]
NBA_df = pd.DataFrame(nba, columns = features)
NBA_df['team']= nba.team
parallel_coordinates(frame=NBA_df,class_column = 'team')

```

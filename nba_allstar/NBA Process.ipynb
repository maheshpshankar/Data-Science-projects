```python
#Join on player name and season
#Create new variable in nba_basic

season_end = [sea for sea in nba_basic.season]
nba_basic['season_end'] = season_end


#Merge function, using inner join....?
all_stats = pd.merge(nba_basic, nba_advance,how = "inner", on = ['player', 'season_end'])
all_stats.info()
```

```python
#Del variables
del all_stats['season_x']
del all_stats['Unnamed: 0_x']
del all_stats['Unnamed: 0_y']
del all_stats['Rk_y']
del all_stats['Rk_x']
del all_stats['Pos']
del all_stats['Age']
del all_stats['MP']
del all_stats['G']
del all_stats['blnk']
del all_stats['blnk2']
#del all_stats['ORB%']
#del all_stats['DRB%']
del all_stats['season_y']
#del all_stats['season_end']
del all_stats['Tm']
```
